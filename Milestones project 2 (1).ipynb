```python
#time series testing for china production yield
import statsmodels.api as sm

china_yield=yield_value.groupby("Area").get_group("China, mainland")

#set Year column to date_time and set the column as index
china_yield["Year"]=pd.to_datetime(china_yield["Year"])
china_yield.set_index("Year", inplace=True)


#perform series regression test
y=china_yield["Value"]
X=range(len(china_yield))
X = sm.add_constant(X) #add a constant term to the predictor
model=sm.OLS(y,X).fit()

print(model.summary())
```

```python
#plt fit plot;A fit plot shows predicted values of the response variable versus actual values of Y.
#If the linear regression model is perfect, the predicted values will exactly equal the observed values 
#and all the data points in a predicted versus actual scatterplot will fall on the 45° diagonal.

sm.graphics.plot_fit(model,1, vlines=True)
plt.title("Fitted Linear Regression-China")
plt.grid(True)
plt.legend()
plt.savefig("china_fitLR.jpg")
```

```python
#linear regression for Vietnam

vietnam_yield=yield_value.groupby("Area").get_group("Viet Nam")

#set Year column to date_time and set the column as index
vietnam_yield["Year"]=pd.to_datetime(vietnam_yield["Year"])
vietnam_yield.set_index("Year", inplace=True)


#perform series regression test
y=vietnam_yield["Value"]
X=range(len(vietnam_yield))
X = sm.add_constant(X) #add a constant term to the predictor
model_viet=sm.OLS(y,X).fit()

print(model_viet.summary())
```

```python
sm.graphics.plot_fit(model_viet,1, vlines=True)
plt.title("Fitted Linear Regression-Vietnam")
plt.grid(True)
plt.legend()
plt.savefig("viet_fitLR.jpg")
```

```python
msia_yield=yield_value.groupby("Area").get_group("Malaysia")

#set Year column to date_time and set the column as index
msia_yield["Year"]=pd.to_datetime(msia_yield["Year"])
msia_yield.set_index("Year", inplace=True)


#perform series regression test
y=msia_yield["Value"]
X=range(len(vietnam_yield))
X = sm.add_constant(X) #add a constant term to the predictor
model_msia=sm.OLS(y,X).fit()

print(model_msia.summary())
```

```python
sm.graphics.plot_fit(model_msia,1, vlines=True)
plt.title("Fitted Linear Regression- Malaysia")
plt.grid(True)
plt.legend()
plt.savefig("msia_fitLR.jpg")
```
